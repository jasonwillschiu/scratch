```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
pytrends
```

```python
kw_list = ['qantas','jetstar','tigerair','air new zealand','virgin australia']
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='AU', gprop='')
```

```python
df = pytrends.interest_over_time()
df
```

```python
df = df.reset_index()
```

```python
df['strdate'] = df.date.dt.strftime('%Y-%m-%d')
```

```python
df['strdate'].tolist()
```

```python
g = hc.Highcharts()

g.plotOptions.area.fillOpacity = 1
g.plotOptions.area.stacking = 'percent'
g.chart.type = 'area'
# g.chart.width = 500
g.chart.height = 800
g.tooltip.enabled = True
# g.yAxis.labels.formatter = hc.scripts.FORMATTER_PERCENT
# g.tooltip.pointFormat = hc.scripts.TOOLTIP_POINT_FORMAT_PERCENT
# g.tooltip.positioner = hc.scripts.TOOLTIP_POSITIONER_CENTER_TOP
g.tooltip.split = True

g.yAxis.title.text = 'Relative search share'

g.title.text = 'Top 5 Airlines in Australia on Google Trends'
g.xAxis.title.text = 'Past 5 years'
g.xAxis.categories = df['strdate'].tolist()
g.series = hc.build.series(df)

g.plot()
```
